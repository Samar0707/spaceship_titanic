```python
# data['GroupId']=data.PassengerId.str
data['GroupId']=data.PassengerId.str.slice(stop=4).astype('float')
data
```

```python
groups = pd.get_dummies(data.GroupId).sum()
groups = pd.DataFrame(groups,columns=["noPassengers"])
groups["inGroup"] = (groups.noPassengers > 1).astype('float')
groups
```

```python
data["inGroup"] = data.GroupId    # Duplicate GroupID
data.inGroup = data.inGroup.replace(groups.inGroup.index,groups.inGroup.values)
data.head(60)
```

```python
groups["HomePlanet"] = np.NaN                                       # Initialize
groups["Cabin"] = np.NaN
groups["Destination"] = np.NaN
groups
```

```python
for x in groups.index:                                              # Iterate through groups
    if groups.noPassengers[x] > 1:                                  # Multiple passengers ?
        GroupId = data.GroupId == x                              # Get indices in dataset
        for y in data.index[GroupId]:# Iterate through passengers
            if pd.notna(data.HomePlanet[y]):                     # Find HomePlanet
                groups.HomePlanet[x] = data.HomePlanet[y]
            if pd.notna(data.Cabin[y]):                          # Find Cabin
                groups.Cabin[x] = data.Cabin[y]
            if pd.notna(data.Destination[y]):                    # Find Destination
                groups.Destination[x] = data.Destination[y]
```

```python
groups
```

```python
GroupId = data.GroupId.copy()
GroupId.replace(groups.HomePlanet.index, groups.HomePlanet.values, inplace=True)
tmp = (pd.isna(data.HomePlanet)) & (data.inGroup == 1)
data.HomePlanet[tmp] = GroupId[tmp]
print("Fixed HomePlanet:     " , tmp.sum())
```

```python
GroupId = data.GroupId.copy()
GroupId.replace(groups.Cabin.index, groups.Cabin.values, inplace=True)
tmp = (pd.isna(data.Cabin)) & (data.inGroup == 1)
data.Cabin[tmp] = GroupId[tmp]
print("Fixed Cabin:          " , tmp.sum())
```

```python
GroupId = data.GroupId.copy()
GroupId.replace(groups.Destination.index, groups.Destination.values, inplace=True)
tmp = (pd.isna(data.Destination)) & (data.inGroup == 1)
data.Destination[tmp] = GroupId[tmp]
print("Fixed Destination:    " , tmp.sum())
```

```python
tmp = (pd.isna(data.Cabin))
data.Cabin[tmp] = 'X/-1/X'
print("Added dummy cabin:    " , tmp.sum())

data["Deck"] = data.Cabin.str.slice(stop=1)                           # Extract Cabin
data["Side"] = data.Cabin.str.slice(start=-1)                         # Extract Side
data["cNum"] = data.Cabin.str.slice(start=2,stop=-2).astype('float')  # Extract Cabin num.

fig = plt.figure(figsize = (16,10))

grp = data.groupby(['Deck','HomePlanet'])['HomePlanet'].size()
grp = grp.unstack().fillna(0).astype('int')
sub = fig.add_subplot(2,4,1)
sns.heatmap(grp, annot=True, linewidths=.5, fmt='d', cbar=False)

grp = data.groupby(['Deck','Destination'])['Destination'].size()
grp = grp.unstack().fillna(0).astype('int')
sub = fig.add_subplot(2,4,2)
sns.heatmap(grp, annot=True, linewidths=.5, fmt='d', cbar=False)

grp = data.groupby(['Side','HomePlanet'])['HomePlanet'].size()
grp = grp.unstack().fillna(0).astype('int')
sub = fig.add_subplot(2,4,3)
sns.heatmap(grp, annot=True, linewidths=.5, fmt='d', cbar=False)
grp = data.groupby(['Side','Destination'])['Destination'].size()
grp = grp.unstack().fillna(0).astype('int')
sub = fig.add_subplot(2,4,4)
sns.heatmap(grp, annot=True, linewidths=.5, fmt='d', cbar=False)

grp = data.groupby(['Destination','HomePlanet'])['HomePlanet'].size()
grp = grp.unstack().fillna(0).astype('int')
sub = fig.add_subplot(2,4,5)
sns.heatmap(grp, annot=True, linewidths=.5, fmt='d', cbar=False)

grp = data.groupby(['Deck','Side'])['Side'].size()
grp = grp.unstack().fillna(0).astype('int')
sub = fig.add_subplot(2,4,6)
sns.heatmap(grp, annot=True, linewidths=.5, fmt='d', cbar=False)

sub = fig.add_subplot(2,4,7)
sns.countplot(data = data, x = 'Deck', palette = "RdYlBu" )

sub = fig.add_subplot(2,4,8)
sns.countplot(data = data, x = 'Side', palette = "RdYlBu" )
```
